# import flipkart dataset

```python
flipkart=pd.read_csv('flipkart_com-ecommerce_sample.csv')
flipkart.head()
```

# data preprocessing steps

```python
flipkart.isnull().sum()
```

```python
flipkart.fillna(method='ffill',inplace=True)#filling empty data using fillna method in pandas
```

```python
flipkart.isnull().sum()#after cleaning no empty data is present
```

```python
flipkart.columns#columns names in the dataset
```

```python
flipkart['product_name'].value_counts()#number of times repeated product_name in flipkart dataset
```

```python
amazon['product_name'].isin(flipkart['product_name']).value_counts()#comparing how many products are same in both amazon and flipkart dataset
```

```python
dataf=flipkart.groupby('product_name').mean().sort_values('retail_price',ascending=False)
dataf#groupby for product_name in flipkart
```

```python
flipkart1=pd.merge(flipkart,amazon,on='image',how='left')
flipkart1#merge flipkart and amazon on left join 
```

```python
flip=flipkart1['product_name_x']#use it later --left joined (product_name) of flipkart
```

```python
amazon1=pd.merge(flipkart,amazon,on='image',how='right')
amazon1#merge flipkart and amazon on right join 
```

```python
ama=amazon1['product_name_y']#use it later --right joined (product_name) of flipkart
```

```python
common_product=pd.merge(dataf,data1,on='product_name',how='inner')
common_product#intersection of both groupby variables of (amazon and flipkart)
```

```python
common_product.reset_index(inplace=True)
common_product#reset product name to column name
```

```python
common_product.columns
```

```python
#removing unwanted columns from common_product data frame
common_product.drop(['is_FK_Advantage_product_x','is_FK_Advantage_product_y','product_name'],axis=1,inplace=True)
```

```python
common_product
```

```python
common_product.insert(0, "product name in flipkart", flip)#inserting  flip data product_name for common_product
```

```python
common_product.insert(3,"product name in amazon",ama)#inserting ama product_name for common_product
```

```python
common_product
```

# renameing column names

```python
common_product.columns=['Product name in Flipkart','Retail Price in Flipkart','Discounted Price in Flipkart','Product name in Amazon','Retail Price in Amazon','Discounter Price in Amazon']
```
